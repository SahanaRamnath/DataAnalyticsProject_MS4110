```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
hap=pd.read_csv('happiness_level.csv').dropna()

```

```python
hap
```

```python
hap['voteDate']=pd.to_datetime(hap['voteDate'],infer_datetime_format=True)
hap['ordinal']=[x.toordinal() for x in hap['voteDate']]
```

```python
max(hap['ordinal'])
```

```python
hap['ordinal']=hap['ordinal']/736408
```

```python
hap
```

```python
hap["unique_emp"] = hap["employee"].map(str) + hap["companyAlias"]
```

```python
hap
```

```python
hap.sort_values(['unique_emp'])
```

```python
empl = hap['unique_emp'].unique()
```

```python
empl.shape
```

```python
happ = np.zeros(4377)
```

```python
index=0
for i in empl:
    temp=0
    a = hap.loc[hap['unique_emp'] == i]
    for j in range(len(a)):
        temp=temp+(a.iloc[j,3]*np.exp(a.iloc[j,4]))
    happ[index]=temp
    index=index+1
        
        
```

```python
happ
```

```python
empl
```

```python
happy_dict={}
index=0
for i in empl:
    happy_dict[i]=happ[index]
    index=index+1
```

```python
np.save('HappinessFeature.npy',happy_dict)
```
